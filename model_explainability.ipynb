```python
import pandas as pd
import shap
import joblib
import matplotlib.pyplot as plt

# Load clean dataset
df = pd.read_csv('Dataset/clean_fraud_dataset.csv')
```

```python
# Features & Target
X = df.drop(columns=['is_fraud', 'trans_date_trans_time', 'merchant', 'city'], errors='ignore')
X = X.select_dtypes(include=['number'])
y = df['is_fraud']

# Load trained model
model = joblib.load('trained_model.pkl')

# SHAP Explainability
explainer = shap.Explainer(model)
shap_values = explainer(X)
```

```python
# Summary plot
shap.summary_plot(shap_values, X)

# Force plot for first prediction
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values.values[0,:], X.iloc[0,:])

print("✅ SHAP Explainability Completed.")
```
