```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load clean dataset
df = pd.read_csv('Dataset/clean_fraud_dataset.csv')
```

```python
# Overall fraud distribution
fraud_counts = df['is_fraud'].value_counts(normalize=True) * 100
print("\nFraud Distribution (%):\n", fraud_counts)
```

```python
# Gender vs Fraud
plt.figure(figsize=(6,4))
sns.countplot(x='gender_M', hue='is_fraud', data=df)
plt.title('Gender vs Fraud Distribution')
plt.show()
```

```python
# State-wise Fraud distribution (Top 10 states by fraud count)
state_fraud = df[df['is_fraud']==1].filter(like='state_', axis=1).sum().sort_values(ascending=False)
state_fraud.head(10).plot(kind='bar', figsize=(8,4), title='Top 10 States by Fraud Count')
plt.show()

print("✅ Demographic Fraud Pattern Analysis Completed.")
```
